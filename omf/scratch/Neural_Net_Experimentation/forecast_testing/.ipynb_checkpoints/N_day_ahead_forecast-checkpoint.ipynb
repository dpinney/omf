```python
import loadForecast as lf
import pandas as pd
```

```python
df = pd.read_csv('data/test/COAST.csv', parse_dates=['dates'])
```

```python
day = [24*i for i in range(1, 11)]
temp = [2.5 + 3*i for i in range(10)]
temp[5:] = [15]*5
```

```python
acc_dict = {}
for i, (d, t) in enumerate(zip(day, temp)):
    all_X = lf.makeUsefulDf(df, hours_prior=d, noise=t)
    all_X = all_X.drop(['load_prev_n'], axis=1)
    all_y = df['load']
    _, acc = lf.neural_net_predictions(all_X, all_y)
    acc_dict[i] = acc
```

```python
%matplotlib inline
d = pd.DataFrame(acc_dict).T
d['test'].plot()
```
